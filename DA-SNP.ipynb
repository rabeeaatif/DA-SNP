```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pylab as pl
from sklearn.preprocessing import LabelEncoder
import xlrd
```

## Pre-Processing

```python
#for the file to load, please upload the data file in the collab 
df = pd.read_excel('/content/S&P Analyst Test Data - Sections A and B (as applicable) (1).xlsx', 0)
```

```python
new_header = df.iloc[2] #grab the first row for the header
df = df[3:] #take the data less the header row
df.columns = new_header #set the header row as the df header
df
```

## Question 1 

```python
#coversion to date time format for yera extraction
df['item_creation_time']= pd.to_datetime(df['item_creation_time'])

#finding all products created in 2018 
subset_df = df[df["item_creation_time"].dt.year == 2018]
subset_df
```

```python
print("unique products that were created in the year 2018 = ", subset_df['itemid'].nunique() )
```

## Question 2

```python
# top 3 categories that have the greatest number of unique cross border products
un = df.loc[df['cb_option'] == 1]
un.drop_duplicates(subset=['itemid'])
un.groupby('category')['cb_option'].count().nlargest(3)
```

## Question 3

```python
# top 3 shops with highest revenue
df['Revenue'] = df['price'] * df['sold_count']
df.groupby('shopid')['Revenue'].sum().astype(float).nlargest(3)
```

## Question 4

```python
# duplicated listings within each shop


df["is_duplicate"] = df.duplicated(subset=['item_name','item_description', 'price', 'shopid'])
#df.loc[df['is_duplicate'] == True]
df
```

## Question 5

```python
# duplicated listings that have less than 2 sold_count also save in a different excel file

result = pd.ExcelWriter('duplicated_listings.xlsx')
duplicated_listings = df.loc[(df['is_duplicate'] == True) & (df['sold_count'] < 2)]
duplicated_listings.to_excel(result)  

result.save()
```
